```python
import copy

def solarize(x):
    return -255 / 2 * np.cos(4 * np.pi * x / 255) + 255/2

solarization = copy.deepcopy(img_gray)

for i in range(len(img_gray)):
    for j in range(len(img_gray[0])):
        solarization[i][j] = solarize(img_gray[i][j])

cv2.imwrite("solarization.jpg", solarization)
```

```python
from tqdm import tqdm

def cvt_solarize(x):
    return -255 / 2 * np.cos(4 * np.pi * x / 255) + 255/2

def solarize_RGB(img):

    a = len(img)
    b = len(img[0])
    c = len(img[0][0])
    bar = tqdm(total = a*b*c)

    solarization_RGB = copy.deepcopy(img)

    for i in range(a):
        for j in range(b):
            for k in range(c):
                solarization_RGB[i][j][k] = cvt_solarize(img[i][j][k])
            bar.update(c)

    beep(1000, 1000)
    return solarization_RGB
```

```python
solarization_RGB = solarize_RGB(img)
cv2.imwrite("solarization_RGB.jpg", solarization_RGB)
```

```python
import os
def beep(freq, dur):
    os.system('play -n synth %s sin %s' % (dur/1000, freq))
```

```python
beep(1000, 1000)
```

```python
# 拡大（中心を支点にする）
import copy
from tqdm import tqdm

def enlarge(img):
    enlargement = copy.deepcopy(img)

    a = len(img)
    b = len(img[0])
    c = len(img[0][0])
    bar = tqdm(total = a*b*c)
    temp = img[772:2316, 579:1737]
    print(temp.shape)
    
    A = len(temp)
    B = len(temp[0])
    C = len(temp[0][0])
    for i in range(A):
        for j in range(B):
            for k in range(C):
                enlargement[i*2][j*2][k] = temp[i][j][k]
                enlargement[i*2+1][j*2][k] = temp[i][j][k]
                enlargement[i*2][j*2+1][k] = temp[i][j][k]
                enlargement[i*2+1][j*2+1][k] = temp[i][j][k]
            bar.update(c*4)

    beep(1000, 1000)
    return enlargement
```

```python
enlargement = enlarge(img)
cv2.imwrite("enlarge.jpg", enlargement)
```

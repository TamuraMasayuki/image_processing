```python
import cv2
import copy
```

```python
img_1 = cv2.imread("tahara_1.jpg")
img_2 = cv2.imread("tahara_2.jpg")
img_3 = cv2.imread("tahara_3.jpg")
img_4 = cv2.imread("tahara_4.jpg")
img_5 = cv2.imread("tahara_5.jpg")
img_6 = cv2.imread("tanaka_1.jpg")
img_7 = cv2.imread("tanaka_2.jpg")
```

```python
img_1 = cv2.cvtColor(img_1, cv2.COLOR_BGR2GRAY)
img_2 = cv2.cvtColor(img_2, cv2.COLOR_BGR2GRAY)
img_3 = cv2.cvtColor(img_3, cv2.COLOR_BGR2GRAY)
img_4 = cv2.cvtColor(img_4, cv2.COLOR_BGR2GRAY)
img_5 = cv2.cvtColor(img_5, cv2.COLOR_BGR2GRAY)
img_6 = cv2.cvtColor(img_6, cv2.COLOR_BGR2GRAY)
img_7 = cv2.cvtColor(img_7, cv2.COLOR_BGR2GRAY)
```

```python
def polarize(img):
    for i in range(len(img)):
        for j in range(len(img[0])):
            if img[i][j] < 128:
                img[i][j] = 0
            else:
                img[i][j] = 255
    return img
```

```python
img_1 = polarize(img_1)
```

```python
img_2 = polarize(img_2)
img_3 = polarize(img_3)
```

```python
import cv2
import copy
```

```python
img = cv2.imread("tahara_1.jpg")
```

```python
img_2 = cv2.imread("tahara_2.jpg")
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_2_gray = cv2.cvtColor(img_2, cv2.COLOR_BGR2GRAY)
```

```python
def polarize(img):
    for i in range(len(img)):
        for j in range(len(img[0])):
            if img[i][j] < 128:
                img[i][j] = 0
            else:
                img[i][j] = 255
    return img
```

```python
img_polarization = polarize(img_gray)
img_2_polarization = polarize(img_2_gray)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img_polarization)
```

```python
plt.imshow(img_2_polarization)
```
